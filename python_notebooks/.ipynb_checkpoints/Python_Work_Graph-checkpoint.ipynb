```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = [104, 235, 348, 354, 455, 596]
df = pd.DataFrame(data, columns=['Salary'])
df
```

```python
plt.pie(df,labels=df['Salary'])
plt.title("Salaries in the office")
plt.show()
```

```python
office_workers=["Jim","Pam","Kevin","Oscar","Dwight","Michael"]
explode = (0, 0, 0, 0,0,0.1)
plt.pie(df,explode= explode, labels=office_workers,  autopct='%1.1f%%', shadow= True)
plt.title("Salaries in the office(%)")
plt.text(1,1,"$2092 in all")
plt.show()
```
